```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib as mpl
import matplotlib.pyplot as plt
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
file1="C:/Users/farha_5s5ygrk/Assignment_4_data/Movie_collection_Independent.csv"
file2="C:/Users/farha_5s5ygrk/Assignment_4_data/Movie_collection_Target.csv"
data_x=pd.read_csv(file1)
data_y=pd.read_csv(file2)
```

```python
print(data_x.shape)
print(data_y.shape)
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(data_x, data_y, random_state=42)
X_train_full, X_valid, y_train_full, y_valid =  train_test_split(data_x, data_y, random_state=42)
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train_full)
X_test_scaled=scaler.fit_transform(X_test)
X_valid_scaled=scaler.fit_transform(X_valid)
```

```python
input_ = keras.layers.Input(shape=X_train_full.shape[1:])
hidden1 = keras.layers.Dense(30, activation="relu")(input_)   #input used as a function
hidden2 = keras.layers.Dense(30, activation="relu")(hidden1)
concat = keras.layers.concatenate([input_, hidden2])
output = keras.layers.Dense(1)(concat)
model = keras.models.Model(inputs=[input_], outputs=[output])
```

```python
model.compile(loss="mean_squared_error", 
              optimizer=keras.optimizers.SGD(learning_rate=1e-1), 
              metrics=['mae'])
history=model.fit(X_train_scaled, y_train_full, epochs=100, validation_data=(X_valid_scaled, y_valid), verbose=1)
```

```python
mae_test = model.evaluate(X_test_scaled, y_test)
```
